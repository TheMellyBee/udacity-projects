# Days Between Dates

This lesson will focus on one problem: calculating the number of days between two dates.

```python

class Date :
    days_in_months = [31, 28, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31]
    isLeapYear = False
        
    def __init__ (self, year, month, day):
        self.year = year
        self.month = month
        self.day = day
        
        self.__check_leap_year()
            
    def __check_leap_year(self):
        if self.year%4 == 0 and self.month < 3:
            self.isLeapYear = True
        return False
            
    def progress_date(self):
        self.day += 1 

        if self.day > self.days_in_months[self.month - 1]:
            if self.isLeapYear and self.month == 2 and self.day == 29:
                self.isLeapYear = False
            else:
                self.day = 1
                self.month += 1 

                if self.month > 12:
                    self.month = 1 
                    self.year += 1
                    
        self.__check_leap_year()
                    
                    
    def __str__(self):
        return "Year: {0} Month: {1} Day: {2}".format(self.year, self.month, self.day)
        
    def __lt__ (self, other):
        if self.year < other.year:
            return True
        elif self.month < other.month:
            return True
        elif self.day < other.day:
            return True
        
        return False
    
    def __eq__ (self, other):
        if self.year == other.year and self.month == other.month and self.day == other.day:
            return True
        return False

    

def daysBetweenDates(year1, month1, day1, year2, month2, day2):
    """
    Calculates the number of days between two dates.
    """
    date1 = Date(year1, month1, day1)
    date2 = Date(year2, month2, day2)
    
    days = 0
    
    if date1 == date2:
        return days
    
    while date1 < date2:
        date1.progress_date()
        days += 1
      
    
    return days

def testDaysBetweenDates():
    
    
    # test same day 0
    assert(daysBetweenDates(2017, 12, 30, 2017, 12, 30)==0)
    
    # test adjacent days 1
    assert(daysBetweenDates(2017, 12, 30, 2017, 12, 31) == 1)
    
    # test new year 2
    assert(daysBetweenDates(2017, 12, 30,2018, 1,  1) == 2)
    
    # test full year difference 365
    assert(daysBetweenDates(2012, 6, 29, 2013, 6, 29) == 365)
    
    assert(daysBetweenDates(2012, 2, 27, 2012, 3, 1) == 3)
    
    print("Congratulations! Your daysBetweenDates")
    print("function is working correctly!")
    
testDaysBetweenDates()
```
